```python
'''
DeepLearning

RNN
ARIMA：基于统计

HMM（隐马尔科夫模型）传统机器学习用HMM
在机器学习里面做时间序列 基本上就是HMM较多
但现在HMM用的比较少，用deeplearning比较多

deeplearn中时序做的比较多的是RNN
LSTM


AR:描述值之间的关系[自回归]
自回归模型描述的是当前值与历史值之间的关系


MA模型：描述误差之间的关系[移动平均]
滑动平均模型描述的是自回归部分的误差累计

'''
'''
ARMA模型要求时间序列是平稳的
如何看时间序列是否是平稳的？
一个时间序列，如果均值没有系统的变化（无趋势）、方差没有系统变化
严格消除了周期性的变化，就是平稳的


均值：平均值
方差：代表波动范围

I差分：做差 
将非平稳的数据转换为平稳的数据
一阶差分：
二阶差分：
 
通常非平稳序列可以经过d阶差分 基本上就可以得到弱平稳或近似平稳的时间序列

d阶差分表示I(d)
  
'''
'''
ARIMA(p,d,q)模型：p阶自回归滞后项，q阶滑动平均滞后项，d阶差分
 
'''
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
#构造时间序列
df_obj = pd.DataFrame(np.random.randn(1000, 1),
                      index=pd.date_range('20170101',periods=1000),
                      columns=['data'])
df_obj['data'] = df_obj['data'].cumsum()
print(df_obj.head())
```

```python
# 一阶差分处理
df_obj['diff_1'] = df_obj['data'].diff(1)
# 二阶差分处理 (二阶差分是在新生成的一阶差分上做差分)
df_obj['diff_2'] = df_obj['diff_1'].diff(1)

#查看图像
df_obj.plot(subplots=True, figsize=(18, 12))
```
