```python
'''
滑动窗口（moving window function）
在时间窗口上计算各种统计函数
窗口函数（window functions ）
滚动统计(rolling)
obj.rolling().func
window 窗口大小
center 窗口是否居中统计
'''
```

```python
import pandas as pd
import numpy as np
ser_obj = pd.Series(np.random.randn(1000),
                   index=pd.date_range('20170101', periods=1000))
ser_obj = ser_obj.cumsum()
print(ser_obj.head())
```

```python
r_obj = ser_obj.rolling(window=5)
print(r_obj)
```

```python
print(r_obj.mean())
```

```python
#画图查看
import matplotlib.pyplot as plt
%matplotlib inline
plt.figure(figsize=(15, 5))
ser_obj.plot(style=r'')
ser_obj.rolling(window=10, center=True).mean().plot(style='b')
```

```python
print(ser_obj.rolling(window=10).mean().plot(style='b'))
```
